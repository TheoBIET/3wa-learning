## Chapitre 4 - fréquence

```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

1. Récupérez les données du dataset dans un fichier Python ou Notebook.

```python
with open ('./students.csv', 'r') as f:
    data_reader = csv.reader(f, delimiter=',', quotechar='"')
    # Supprimer tous les passagers qui n'ont pas d'âge
    data = [(d[0], d[1], d[2], d[3], d[4])for d in data_reader][1:]
    
data = np.array(data, dtype=np.dtype([
    ('name', np.dtype('U14')), 
    ('eye_color', np.dtype('U8')),
    ('sex', np.dtype('U2')),
    ('mention', np.dtype('U2')),
    ('note', np.dtype('f8'))
]))

total_student = data.size

print(data[:5])
```

2. Edutiez le critère "Couleur des yeux" et calculez la fréquence d'apparition de ce critère.

```python
colors = np.unique(data['eye_color'])
nbColors = colors.size
frequences = np.empty(nbColors, dtype=np.dtype('f8'))

fColor = lambda color: np.around((data['eye_color'] == color).sum() / total_student, 2)

for i,c in enumerate(colors):
    frequences[i] = fColor(c)
    
print(frequences)
```

3. Créez à partir du tableau précédent un diagramme en secteur pour visualiser la répartition de la couleur des yeux.

```python
plt.figure(figsize=(12,8))

plt.pie(frequences, explode=(0.05, 0.05, 0.05, 0.05), labels=['Bleu', 'Marron', 'Noir', 'Vert'], autopct='%1.1f%%',
        shadow=True, startangle=90, colors=['#a1caf1', '#603101', '#2e2b2e', '#337a2c'])

plt.title('Fréquence de la couleur des yeux') 

plt.show()
```
